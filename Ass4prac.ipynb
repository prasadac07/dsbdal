```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df["age"]=df["age"].fillna(df["age"].median())
```

```python
df.isnull().sum()
```

```python
df["embarked"]=df["embarked"].fillna(df["embarked"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["embark_town"]=df["embark_town"].fillna(df["embark_town"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.drop(columns=["deck"],inplace=True)
```

```python
df.shape
```

```python
#1.Histogram
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=df,x="fare",bins=20,kde=True,color="blue")
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Number of passengers")
plt.show()
```

```python
#2.Countplot
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df,x="class",hue="survived")
plt.title("Survived by class")
plt.xlabel("Class")
plt.ylabel("Number of passengers")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df,x="sex",hue="survived",palette="Set1")
plt.title("Survived by Gender")
plt.xlabel("gender")
plt.ylabel("Number of passengers")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="survived", y="fare")
plt.title('Ticket Fare Distribution by Survival Status')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.xticks([0, 1], ['Died', 'Survived'])
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x="survived", y="fare")
plt.title('Ticket Fare Distribution by Survival Status')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.xticks([0, 1], ['Died', 'Survived'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df,x="age",y="fare",line_kws={"color":"red"})
plt.title("Reg plot")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
numerical_corr=df[["age","fare","parch","sibsp"]].corr()
sns.heatmap(data=numerical_corr,annot=True,cmap="coolwarm")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
