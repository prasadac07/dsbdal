```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of features
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Create a 2x2 grid for subplots
plt.figure(figsize=(10, 8))

# Loop through the features and plot KDE plots
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)  # Automatically positions the plot in a 2x2 grid
    sns.kdeplot(data=df, x=feature, hue="species", fill=True)
    plt.title(f"KDE of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Density")

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns

# Load the Iris dataset
df = sns.load_dataset('iris')

# Define the function to remove outliers using IQR for multiple features
def remove_outliers_multiple(features):
    global df
    for feature in features:
        # Calculate the first and third quartiles (Q1 and Q3)
        q1, q3 = np.percentile(df[feature], [25, 75])
        
        # Calculate the Interquartile Range (IQR)
        iqr = q3 - q1
        
        # Define the lower and upper bounds to identify outliers
        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr
        
        # Filter the dataframe to remove outliers
        df = df[(df[feature] >= lower) & (df[feature] <= upper)]

# List of features to check for outliers
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Call the function to remove outliers
remove_outliers_multiple(features)

# Check the shape of the dataset after outlier removal
print(f"Shape of the dataset after removing outliers: {df.shape}")

# Verify by plotting boxplots for each feature again to check outliers
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, x="species", y=feature)
    plt.title(f"Boxplot of {feature} after outlier removal")
    plt.xlabel("Species")
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
df = sns.load_dataset('iris')

# Define the function to remove outliers using IQR for multiple features
def remove_outliers_multiple(features, multiplier=1.5):
    global df
    for feature in features:
        # Calculate the first and third quartiles (Q1 and Q3)
        q1, q3 = np.percentile(df[feature], [25, 75])
        
        # Calculate the Interquartile Range (IQR)
        iqr = q3 - q1
        
        # Define the lower and upper bounds to identify outliers
        lower = q1 - multiplier * iqr
        upper = q3 + multiplier * iqr
        
        # Filter the dataframe to remove outliers
        df = df[(df[feature] >= lower) & (df[feature] <= upper)]

# List of features to check for outliers
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Call the function to remove outliers with multiplier 3.0
remove_outliers_multiple(features, multiplier=3.0)

# Check the shape of the dataset after outlier removal
print(f"Shape of the dataset after removing outliers: {df.shape}")

# Verify by plotting boxplots for each feature again to check outliers
plt.figure(figsize=(12, 10))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, x="species", y=feature)
    plt.title(f"Boxplot of {feature} after outlier removal")
    plt.xlabel("Species")
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```
