```python
import seaborn as sns
import pandas as pd

# Load the Iris dataset using seaborn
df = sns.load_dataset('iris')
df.head()
```

```python
df["species"].value_counts()
```

```python

# Display basic statistics (mean, std, etc.) for each species
species_group = df.groupby('species')

# Calculate basic statistics
statistics = species_group.describe().T  # Transpose to make it more readable

# Calculate specific percentiles (e.g., 25%, 50%, and 75%)
percentiles = species_group.quantile([0.25, 0.5, 0.75])

# Display the results
print("Basic Statistics for Each Species:\n")
print(statistics)
print("\nPercentiles (25%, 50%, 75%) for Each Species:\n")
print(percentiles)
```
